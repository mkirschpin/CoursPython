```python
import pandas as pnd
```

```python
visits = pnd.read_csv('http://kirschpm.fr/cours/PythonDataScience/datasets_projet/csv/DoctorVisits.csv', delimiter=',',  header=[0],index_col=[0])
visits.info()
```

```python
visits.head()
```

On souhaite créer une seule colonne 'assurance' qui indique s'il y a une assurance 'private' (valeur 0), une assurance 'freepoor' (valeur 1), une assurance 'freerepat' (valeur 2), ou aucune des 3 (valeur 3)
On crée une série appelée nv_val, à laquelle on va donner la valeur 0, 1, 2 ou 3 selon la valeur des colonnes 7 ('private'), 8 ('freepoor') et 9 ('freerepat').

Ensuite on ajoute une nouvelle colonne au dataframe, que l'on appelle 'assurance' et à laquelle on affecte les valeurs de la série nv_val. 

Pour créer notre série nv_val, on va utiliser une petite boucle *for*, qui va parcourir tous les lignes de notre DataFrame visits. Pour chacune de ces lignes, on regarde la valeurs des colonnes 7 ('private'), 8 ('freepoor') et 9 ('freerepat'). Afin de récupérer la valeur qu'on retrouve dans ces colonnes, on va utiliser l'opération *iat[lin,col]* qui récupére la valeur se trouvant à la ligne *lin* et à la colonne *col* du DataFrame.    

```python
import numpy as np
nv_val = []
for i in np.arange(len(visits.index)):
    if visits.iat[i,7] == 'yes':
        nv_val.append('0')
    elif visits.iat[i,8] == 'yes':
        nv_val.append('1')
    elif visits.iat[i,9] == 'yes':
        nv_val.append('2')
    else:
        nv_val.append('3')
print("nv_val : ", nv_val)
```

```python
visits['assurance'] = nv_val
print("df : \n", visits)
```

```python
visits.info()
```

Si on vérifie avec info, la nouvelle colonne est indiquée comme étant "*object*", on va donc ajuster son type avec ***to_numeric*** afin qu'elle soit reconnue comme un numéro entier.  

```python
visits['assurance'] = pnd.to_numeric(visits['assurance'])
visits.info()
```
